## Rules of regex processing
1. The character ^ means start and $ menas end.

2. The character '^' before a string means that the texte regex processor retrieves must start with the string you specify.

3. If you put the $ after the string, it means that the text regex retrieves must end with the string you specify


```python
re.search("^Amy", text) # returns a re.Match object
```

## Patterns and Character Classes

```python
grades="ACAAAABCBCBAA"

print(re.findall("B", grades)) # Return the occurences of B in grades
print(re.findall("AB", grades)) # Return A followed by B
print(re.findall("[AB]", grades)) # Return the occurences of A's and B's, but not necesseraly one followed by the other
```

```python
re.findall("[A][B-C]", grades) # Instances where student receive an A followed by a B or C
```

[AB] pattern describes a set of possible characters which could be either (A OR B), while the [A][B-C] pattern denoted two sets of characters which must have been matched back to back.

```python
re.findall("AB|AC",grades)
```

```python
re.findall("[^A]", grades) # Return the occurences of patterns different of whats after the character ^, because it's written inside the brackets
```

```python
re.findall("^[^A]", grades) # As it wants to match any value at the beginning of the string which value is not an A, it returns an empty list
```

```python
re.findall("^AC$", grades)
```

## Quantifiers
Quantifiers are the number of times you want a pattern to be matched in order to match. The most basic is expressed as e{m,n} where e is the expression we are matching, m is the minimun number of times you want it to be matched and n is the maximum number of times the item could be matched.

```python
# Example. How many times has this student grades been on a back-to-back A's streak
re.findall("A{2,10}", grades) # Minimum of 2 A's in a row and max of 10 (just so it's big enough)
```

```python
re.findall("A{1,1}A{1,1}", grades)
```

```python
# Using this, we could find a decreasing trend in a student's grades
re.findall("A{1,10}B{1,10}C{1,10}", grades)
```

## Groups
Regex is a single pattern which is matched. But you can actually match different patterns(groups) at the same time and the refer to the groups you want.

To group patterns you use parentheses

```python
re.findall("([\w ]*)(\[edit\])",wiki)
```

```python
# Now, how do we get back a list of match objects? Function finditer()
for item in re.finditer("([\w ]*)(\[edit\])", wiki):
  print(item.groups())
```

```python
for item in re.finditer("([\w ]*)(\[edit\])", wiki):
  print(item.group(0))
  print(item.group(1))
```

```python
# You can also label the groups and look at the results as a dictionary
# The syntax is (?P<name>), ehere the parentheses starts the group, the ?P indicates that is
# an extension to basic regexes and <name> is the dictionary key we want to use wrapped in <>

for item in re.finditer("(?P<title>[\w ]*)(?P<edit_link>\[edit\])",wiki):
  print(item.groupdict()['title'])
```

```python
print(item.groupdict())
```

## Look-ahead and Look-behind
In this case, the
pattern being given to the regex engine is for text either before or after the text we are trying to isolate. For example, in our headers we want to isolate text which  comes before the [edit] rendering, but
we actually don't care about the [edit] text itself. Thus far we have been throwing the [edit] away, but if we want to use them to match but don't want to capture them we could put them in a group and use look ahead
instead with ?= syntax

```python
for item in re.finditer("(?P<title>[\w ]+)(?=\[edit\])", wiki):
  # We are matching two groups. The first named title and will have any amount of whitespace
  # or regular characters, the second will be the characters [edit] but we don't actually want this
  # edit put in our output match objects

  print(item)
```

```python
import re
def grades():
    with open ("assets/grades.txt", "r") as file:
        grades = file.read()

    pattern = "(?P<name>[A-Z].*):\sB"
    names=[]
    for person in re.finditer(pattern,grades):
        names.append(person.groupdict()['name'])
    return names

names = grades()
len(names)
```

```python
import re
def logs():
    with open("assets/logdata.txt", "r") as file:
        logdata = file.read()
    pattern = """
   (?P<host>\d+\.\d+\.\d+\.\d+)
   (\s[-]\s)
   (?P<user_name>\w+|[-]+)
   (\s\[)
   (?P<time>\d+\/\w+\/\d+:\d+:\d+:\d+\s-\d+)
   (\]\s")
   (?P<request>.*)
   (".*)
   """
    result = []
    for item in re.finditer(pattern, logdata, re.VERBOSE):
        result.append(item.groupdict())
    return result

result = logs()
result
```
